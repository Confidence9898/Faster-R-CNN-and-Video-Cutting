```python
import pandas as pd
import xml.etree.ElementTree as ET
import os
import glob

train_df = pd.DataFrame(columns=['FileName', 'XMin', 'XMax', 'YMin', 'YMax', 'ClassName'])

# data_path = "Dataset/train/annotation"
# annotation_data = os.listdir(data_path)

# print(annotation_data)
# print(annotation_data)
for data in glob.glob("Dataset/train/annotation/*.xml"):
    tree = ET.parse(data)
    root = tree.getroot()
    print("1")
    # retrieve all <bndbox>
    bndbox_values = []
    filename = root.find('filename').text
    print(filename)
    name = root.find('object/name').text
    if name is None:
        continue
    print(name)
    for bndbox in root.findall('object/bndbox'):
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)
        bndbox_values.append([xmin,ymin,xmax,ymax])

    values = [[filename, bndbox_values[0][0], bndbox_values[0][2], bndbox_values[0][1], bndbox_values[0][3], name]]

    train_df = pd.concat([train_df, pd.DataFrame(values, columns=['FileName', 'XMin', 'XMax', 'YMin', 'YMax', 'ClassName'])], ignore_index=True)
    print(train_df)
train_df.to_csv("train.csv")
```

```python
import pandas as pd
import os

train_df = pd.read_csv('train.csv')

# for training
with open("annotation.txt", "w+") as f:
  for idx, row in train_df.iterrows():
    #   print(idx)
    #   print("--------------------------------------------------")
    #   print(row)
      # height = 1080
      # width = 1920
      x1 = int(row['XMin'])
      x2 = int(row['XMax'])
      y1 = int(row['YMin'])
      y2 = int(row['YMax'])
      
      google_colab_file_path = 'drive/MyDrive/Human_Detection/Dataset/train/images/'
      fileName = os.path.join(google_colab_file_path, row['FileName'])
      className = row['ClassName']
      f.write(fileName + ',' + str(x1) + ',' + str(y1) + ',' + str(x2) + ',' + str(y2) + ',' + className + '\n')
```
